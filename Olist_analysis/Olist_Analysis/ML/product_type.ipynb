```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# 读取CSV文件
data = pd.read_csv("C:/Users/KK Chan/Desktop/作業/fianl_project/data/olist_products_dataset.csv")
```

```python
data.product_category_name.value_counts()
```

```python
data.dropna(inplace=True)
```

```python
# 聚類的特征
features = ["product_name_lenght", "product_description_lenght", "product_weight_g", "product_length_cm", "product_height_cm", "product_width_cm"]
X = data[features]
```

```python
# 實例化K-means模型
kmeans = KMeans(n_clusters=50)  # 设置聚類50
kmeans.fit(X)
```

```python
# 獲取每個樣本的聚類標籤
labels = kmeans.labels_
```

```python
# 創立字段["cluster_label"] 加入標籤
data["cluster_label"] = labels
```

```python
# 按排序輸出標籤的聚類結果總數
cluster_counts = data["cluster_label"].value_counts()
cluster_counts_sorted = cluster_counts.sort_index()
cluster_counts_sorted
```

```python
# 輸出現有的每個分類中商品 的聚類分析結果
groubyed_df =  data.groupby('product_category_name')["cluster_label"].value_counts()
```
